```python
"""QUESTÃO 1"""

import pandas as pd
import numpy as np

Mercearia = pd.DataFrame({"loja":["A", "A", "A"],
"dia": ["sex", "sab", "dom"],
"valor": [7500, 9500, 8200]})

Farmacia = pd.DataFrame({"loja":["B", "B", "B"],
"dia": ["sex", "sab", "dom"],
"valor": [5100, 8250, 9900]})

Restaurante = pd.DataFrame({"loja":["C", "C"],
"dia": ["sab", "dom"],
"valor": [7500, 11800]})

"""Concatenação no sentido da coluna"""
pd.concat([Mercearia, Farmacia, Restaurante], ignore_index=True)
```

```python
"""Concatenação no sentido da linha"""

"""
Concatenar dois ou mais DataFrames pandas consiste na operação de empilhar um "no topo" do 
outro, gerando um DataFrame único como resultado.

A função concat do pandas por padrão empilha um por baixo do outro, então, se você criar o concat no sentido da coluna
ele ira concatenar os três dataFrames -Mercearia, Farmacia e Restaurante- empilhados verticalmente.
No caso quando se coloca a opção "axis=1", para o sentido da linha, ele irá concatenar da esquerda para a 
direita, e no caso do Restaurante -opção C-, que funcionou apenas na sexta e no sábado, ele irá apresentar no último índice
valores NaN.
"""

pd.concat([Mercearia, Farmacia, Restaurante], axis=1)
```

```python
"""QUESTÃO 2"""


SQL = pd.DataFrame({"email":["rakesh@xyz.com",
"ecg@acmecorpus.com"]})

PYTHON = pd.DataFrame({"email":["ana@xyz.com",
"jonas@acmecorpus.com",
"rakesh@xyz.com"]})

# A) quais são os alunos distintos, considerando ambas as listas?
"""
Primeiramente foi feita a concatenação dos dois dataFrames, posteriormente, utilizei a função "drop_duplicates()"
para retirar os valores duplicados, ou seja, deixar apenas os alunos diferentes em ambos os dataFrames.
"""
juncao = pd.concat([SQL, PYTHON], ignore_index=True)
alunos_distintos = juncao.drop_duplicates()
alunos_distintos
```

```python
# B) Quais são os alunos que fizeram ambos os cursos?

"""
Utilizei o merge como os dataFrames são compatíveis, para reproduzir uma interseção
e imprimir os alunos que fizeram ambos os cursos.
"""
inter = SQL.merge(PYTHON)
inter
```

```python
# C) quais os alunos que realizaram o curso de SQL, mas não o de PYTHON (e vice-versa)?

apenas_sql = SQL[SQL.email.isin(PYTHON.email) == False]
apenas_python = PYTHON[PYTHON.email.isin(SQL.email) == False]

"""
Nessa questão foi utilizado o comando isin(), ele vai devolver como True os valores iguais que estão
no apenas_sql e no apenas_python, utilizer o == False, para devolver apenas os valores distintos dos cursos,
ou seja, os alunos diferentes que realizaram os cursos.
"""

print("Apenas o SQL: \n{}".format(apenas_sql))

print("\nApenas o Pyhton: \n{}".format(apenas_python))
```
