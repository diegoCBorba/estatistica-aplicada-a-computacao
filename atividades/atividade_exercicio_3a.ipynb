```python
"""QUESTÃO 4"""

import pandas as pd
import numpy as np

df2 = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df2.rename(columns={'Type': 'CarType'}, inplace = True)

chave = list(df2.keys())
tamanho = len(df2)

dicionario = {}

for i in range(tamanho):
    try:
        nome = chave[i].replace('.', '_')
    except:
        continue
    else:
        dicionario[chave[i]] = nome
        
df2.rename(columns= dicionario).head()
```

```python
"""QUESTÂO 6"""
import pandas as pd
Z = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

df_refatorado = Z.isnull().sum().sort_values(ascending=False) # Retornar a soma dos valores nulos encontrados em ordem decrescente
# onde o primeiro valor já representa a coluna de maior valor.

print("A soma dos valores ausentes são {}". format(sum(df_refatorado)))
print("A coluna com o valor máximo ({}) é {}".format(max(df_refatorado), df_refatorado.idxmax()))
```
