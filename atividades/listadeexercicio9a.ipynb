```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df['Survived'] = df['Survived'].map({1: 'Sim', 0: 'Não'})
df['Embarked'] = df['Embarked'].map({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'})
df['Sex'] = df['Sex'].map({'male': 'Homem', 'female': 'Mulher'})
df['Pclass'] = df['Pclass'].map({1: '1ª classe', 2: '2ª Classe', 3: '3ª Classe'})
df = df.rename(columns={'Survived': 'Sobreviventes', 'Sex': 'Sexo', 'Embarked': 'Embarque', 'Pclass': 'Classe da passagem'})
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.subplot(2, 1, 1) 
df["Embarque"].value_counts().plot.pie(title="Embarque")
plt.ylabel('') 
plt.subplot(2, 1, 2)
df["Embarque"].value_counts().plot.barh()
plt.show()
df["Embarque"].value_counts().plot.bar()
```

```python
%matplotlib inline
plt.subplot(2, 1, 1) 
df["Sobreviventes"].value_counts().plot.pie(title="Sobreviventes")
plt.ylabel('') 
plt.subplot(2, 1, 2)
df["Sobreviventes"].value_counts().plot.barh()
plt.show()
df["Sobreviventes"].value_counts().plot.bar()
```

```python
%matplotlib inline
plt.subplot(2, 1, 1) 
df["Sexo"].value_counts().plot.pie(title="Sexo")
plt.ylabel('') 
plt.subplot(2, 1, 2)
df["Sexo"].value_counts().plot.barh()
plt.show()
df["Sexo"].value_counts().plot.bar()
```

```python
%matplotlib inline
plt.subplot(2, 1, 1) 
df["Classe da passagem"].value_counts().plot.pie(title="Classe da passagem")
plt.ylabel('') 
plt.subplot(2, 1, 2)
df["Classe da passagem"].value_counts().plot.barh()
plt.show()
df["Classe da passagem"].value_counts().plot.bar()
```
